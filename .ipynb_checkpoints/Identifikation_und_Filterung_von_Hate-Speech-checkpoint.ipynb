# Identifikation und Filterung von Hate-Speech auf Sozialen Medien
Ziel des folgenden Projekts ist es aus Daten, die Sozialen Medien entnommen wurden, Hassrede zu erkennen und auszugeben.  

```python
import spacy
nlp = spacy.load("de_core_news_sm")
#erstellen einer Liste in der typische Beleidigungen oder Schimpfw√∂rter aufgelistet werden 
hassrede=["Idioten","St√ºmper","linksversifft","dumm","dummen","h√§√üliche" "h√§√ülich","Idiot","Depp","Trottel","Penner","Schwein","Ratte","Ungeziefer","Wichser","Hurensohn","Fotze","Arschificker","Arsch","Islamisierung","Arschloch","Schlitzauge","Spast","Missgeburt"]
#definieren von Listen in denen die gefilterten Tweets (filtered) und die Tweets in denen keine Hassrede erkannt wurden (approved) sortiert werden 
filtered=[]
approved=[]

#√∂ffnen der zu analysierenden Datei
textin = open('germeval2018.training.txt', 'r', encoding='utf-8')
lines = textin.readlines()
textin.close() 

#Datei wird line f√ºr line und token f√ºr token analysiert
for line in lines:
    doc=nlp(line)
    for token in doc:
        #im Prinzip wird hier jeder Tweet in die approved Liste hinzugef√ºgt
        if token.text not in hassrede and line not in approved:
            approved.append(line)
        #wenn sich ein token aus einer line in der Liste Hassrede befindet dann wird diese line aus approved entfernt und zu filtered hinzugef√ºgt
        elif token.text in hassrede and line not in filtered:
            filtered.append(line)
            if line in approved:
                approved.remove(line)
                break
            
            
#print("Frei von Hassrede:\n" + '\n'.join(approved))
print("\nHassrede erkannt:\n" + '\n'.join(filtered))
```

## Erweiterte Erkennung
Abgesehen von Schimpfw√∂rtern gibt es auch Angriffe und Beleidigungen die nicht nur an einem Wort erkennbar sind. Um Tweets in denen solche enthalten sind finden zu k√∂nnen habe ich zwei Listen hinzugef√ºgt. Erst wenn Inhalte aus beiden Listen in dem Tweet vorhanden sind wird Hassrede erkannt. Es bleiben leider trotzdem immer noch Tweets √ºbrig die nur Anhand eines im Tweet nicht vorhandenden Kontexts als Hassrede erkannt werden k√∂nnten. 

```python
hassrede=["Idioten","Islamisierung","St√ºmper","Hetze","linksversifft","Dumm","dumm","dummen","h√§√üliche" "h√§√ülich","Idiot","Depp","Trottel","Penner","Schwein","Ratte","Ungeziefer","Wichser","Hurensohn","Fotze","Arschificker","Arsch","Arschloch","Schlitzauge","Spast","Missgeburt"]
filtered=[]
approved=[]
#zus√§tzliche Listen bei denen aus kombi1 und kombi2 W√∂rter vorhanden sein m√ºssen um Hassrede zu erkennen
kombi1=["Juden","Fl√ºchtlinge","Moslems"]
kombi2=["Abschieben","Davonjagen","Heimschicken","Invasoren","Schweine"]
kombi3=[]

#√∂ffnen der zu analysierenden Datei
textin = open('germeval2018.training.txt', 'r', encoding='utf-8')
lines = textin.readlines()
textin.close() 

#Datei wird line f√ºr line und token f√ºr token analysiert
for line in lines:
    doc=nlp(line)
    for token in doc:
        #im Prinzip wird hier jeder Tweet in die approved Liste hinzugef√ºgt
        if token.text not in hassrede and line not in approved:
            approved.append(line)
        #wenn sich ein token aus einer line in der Liste Hassrede befindet dann wird diese line aus approved gel√∂scht und zu filtered hinzugef√ºgt
        if token.text in hassrede and line not in filtered:
            filtered.append(line)
            if line in approved:
                approved.remove(line)
                break
        #sollte sich ein token in der Liste kombi1 befindet wird die line zur liste kombi3 hinzugef√ºgt 
        if token.text in kombi1:
            kombi3.append(line)
    for token in doc:
        #wenn ein token aus einer line die sich in kombi3 befindet ebenfalls in kombi2 vorkommt wird die line zu filtered hinzugef√ºgt und aus approved entfernt 
        if token.text in kombi2 and line in kombi3:
            filtered.append(line)
            if line in approved:
                approved.remove(line)
                break
                    
            
            
#print("Frei von Hassrede:\n" + '\n'.join(approved))
print("\nHassrede erkannt:\n" + '\n'.join(filtered))
```
